```r
#Multiple line plots
set.seed(123)
n<-50
x<-rnorm(n)
e1<-rnorm(n);y1<-2+4*x+e1
e2<-rnorm(n);y2<-3+x+e2
plot(c(x,x), c(y1,y2), type = "n", xlab="x", ylab="y") # setting up coord. system
points(x, y1, col = "red", pch="1")
points(x, y2, col = "blue", pch="2")
abline(a=2, b=3, col = "red", lwd=3)
abline(a=3, b=2, col = "blue", lwd=3)
abline(lm(y1~x), col = "red", lty=2)
abline(lm(y2~x), col = "blue", lty=2)
abline(lm(c(y1,y2)~c(x,x)), col="green", lty=2, lwd=3)
title(main="Multiple line and point plot, by (Juber Rahman)")
```

```r
#Recreate both graphs in ggplot2.
#Multiple line plots
library(ggplot2)
set.seed(123)
n<-50
x<-rnorm(n)
e1<-rnorm(n);y1<-2+4*x+e1
e2<-rnorm(n);y2<-3+x+e2
df<-data.frame(x,y1,y2)
p<-ggplot(df,aes(x=x))+geom_point(aes(y=y1),col = "red", pch="1")+geom_point(aes(y=y2),col = "blue", pch="2")+xlab("x")+ylab("y")
p1<-p+geom_abline(intercept=2, slope=3, col = "red", lwd=1)+geom_abline(intercept=3, slope=2, col = "blue", lwd=1)+ggtitle("Multiple line and point plot, by (Juber Rahman)")
p2<-p1+geom_smooth(aes(y = y1), colour="red",linetype="dashed",fullrange=TRUE,se=FALSE,method=lm)+geom_smooth(aes(y = y2), colour="blue",fullrange=TRUE,linetype="dashed",se=FALSE,method=lm)
p2+geom_smooth(aes(y = (y1+y2)/2), colour="green",linetype="dashed",fullrange=TRUE,lwd=1,se=FALSE,method=lm)+theme_bw()+theme(plot.title = element_text(face="bold"),panel.grid.major = element_blank(), panel.grid.minor = element_blank())
```
