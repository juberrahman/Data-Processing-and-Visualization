```r
library(ggplot2)
ggplot(faithful, aes(x=eruptions)) +
geom_density()
```

```r
#add histogram with density plot
ggplot(faithful, aes(x=eruptions,y=..density..)) +
geom_histogram(fill="grey60",colour="red")+geom_density(colour="blue")
```

```r
#group by long /short eruptions
faithful$long <- faithful$eruptions > 3
ggplot(faithful, aes(x=eruptions,y=..density..,fill=long)) +
geom_histogram(colour="black")+geom_density(alpha=.5,size=1)+ scale_fill_discrete(labels=c("short\neruption","long\neruption"))+ ggtitle("Combining histogram/density plot \n grouped by long/short eruptions")
```

```r
# use facet 
ggplot(faithful, aes(x=eruptions,y=..density..,fill=long)) +
geom_histogram(colour="black")+geom_density(alpha=.5,size=1)+ scale_fill_discrete(name="long eruption?",labels=c("FALSE","TRUE"))+ ggtitle("Histogram/density plot with facet_grid()")+facet_grid(.~long)
```

```r
ggplot(faithful, aes(x=long, fill=long)) +
geom_bar()
```

```r
# do axis reverse and add annotation
c0 <- table(faithful$long)[[1]]
c1 <- table(faithful$long)[[2]]
n <- c0+c1
ggplot(faithful, aes(x=long, fill=long)) +geom_bar()+coord_flip()+theme(legend.position="bottom")+
annotate("text",x=FALSE,y=n/2,label=paste("total eruptions(n)=",n))
```

```r
ggplot(faithful, aes(x=long, fill=long)) +
geom_bar()
c0 <- table(faithful$long)[[1]]
c1 <- table(faithful$long)[[2]]
n <- c0+c1
ggplot(faithful, aes(x=long, fill=long)) +
geom_bar() +
annotate("text", x=FALSE, y=n/2, label=paste("total eruptions (n)=", n)) +
coord_flip() +
theme(legend.position="bottom")
print(n)
```

```r
# annotate the histograms
ggplot(faithful, aes(x=long, fill=long)) +geom_bar()+annotate("text",x=FALSE,y=c0/2,label=paste("short eruptions (c0)\nCount=",c0))+
annotate("text",x=TRUE,y=c1/2,label=paste("long eruptions (c0)\nCount=",c1))
```

```r
se=(c0*c1/n)
ggplot(faithful, aes(x=long, fill=long)) +
geom_bar()+
annotate("text",x=FALSE,y=.95*n,label="error bound=count+SE")+
annotate()+
annotate("text",x=FALSE,y=c0/2)+
annotate("text",x=TRUE,y=c1/2))
```

```r
library(ggplot2)
faithful$long <- faithful$eruptions > 3
c0 <- table(faithful$long)[[1]]
c1 <- table(faithful$long)[[2]]
n <- c0+c1
se=(c0*c1/n)
ggplot(faithful, aes(x=long, fill=long)) +
geom_bar()+
annotate("segment",x=FALSE,xend=FALSE,y=c0-se,yend=c0+se,arrow=arrow(ends="both",angle=45,length=unit(.2,"cm")))+
annotate("segment",x=TRUE,xend=TRUE,y=c1-se,yend=c1+se,arrow=arrow(ends="both",angle=45,length=unit(.2,"cm")))+
annotate("text",x=FALSE,y=Inf,vjust=1,label="error bound=count+/-SE")+
annotate("text",x=FALSE,y=Inf,vjust=3,label=paste("SE=(c0*c1/n)=",round(se,2)))+
theme(legend.position="top")
```
