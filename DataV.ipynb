```r
library(ggplot2)
df2 <- data.frame(x = 1:5 , y = 1:25, z = 1:25) 
s <- ggplot(df2, aes(x, y)) 
s+geom_point(aes(shape = z), size = 4) + scale_shape_identity() +geom_text(aes(label=z),hjust=-1,vjust=-1)+ggtitle("Problem 1 : plot of 25 symbols\nby(Juber Rahman)")+theme_classic()+theme(axis.ticks=element_blank(),axis.text = element_blank(),axis.title = element_blank())

```
